```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
import re 
import pandas as pd 
import requests
import random
```

```python
base_url = "https://www.seek.com.au"

def fetch_pages(section_url):
    html = urlopen(section_url).read()
    soup = BeautifulSoup(html, "lxml")
    job = soup.find("div", "_365Hwu1")
    category_links = [base_url + dd.a["href"] for dd in job.findAll("article")]
    return category_links
```

```python
# search all data scientist jobs in australia 
ds_url = "https://www.seek.com.au/data-scientist-jobs/in-All-Australia?"
```

```python
# search all available pages for DS jobs
ds_jobs = []
for number in range(0,12,1):
    page = ds_url + str(number)
    ds_jobs.append(page)

DS_jobs = [fetch_pages(page) for page in ds_jobs]
DS_jobs
```

```python
# machine learning jobs 
ml_url= "https://www.seek.com.au/machine-learning-jobs/in-All-Australia?page="

# Machine Jobs

ml_jobs = []
for number in range(0,3,1):
    page = ml_url + str(number)
    ml_jobs.append(page)

ML_jobs = [fetch_pages(page) for page in ml_jobs]
ML_jobs
```

```python
ba_url = "https://www.seek.com.au/business-analyst-jobs/in-All-Australia?page="

# search all available pages for DS jobs
ba_jobs = []
for number in range(0,87,1):
    page = ba_url + str(number)
    ba_jobs.append(page)

BA_jobs = [fetch_pages(page) for page in ba_jobs]
BA_jobs
```

```python
bd_url = "https://www.seek.com.au/big-data-jobs/in-All-Australia?page="

# Big Data jobs
bd_jobs = []
for number in range(0,8,1):
    page = bd_url + str(number)
    bd_jobs.append(page)

BD_jobs = [fetch_pages(page) for page in bd_jobs]
BD_jobs
```

```python
# join searchs for data accumulation purposes

all_jobs = DS_jobs + ML_jobs + BA_jobs + BD_jobs

# organise list of job pages

all_jobs_final = [item for sublist in all_jobs for item in sublist]
all_jobs_final
```
